```python
from sklearn.datasets import make_multilabel_classification
from sklearn.model_selection import train_test_split
from sklearn.datasets import make_blobs
from models import Dim
```

# Single Label Dataset

```python
# Create and split dataset
X, y = make_blobs(1000, n_features=300, centers=20,
                  cluster_std=2, random_state=34)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=33)

# Initialize dim. red. framework
dim = Dim()
dim.X_train = X_train
dim.y_train = y_train
dim.X_test = X_test
dim.y_test = y_test
```

```python
# Obtain 2-dimensional spaces
dim.new_dim = dict()
dim.apply_dim(num_dim=[2], tflag=[3, 6, 7, 8])

# Plot spaces
dim.plot_artificial(n_rows=2, n_cols=4, figsize=(15, 8), save_name='plot_artificial_24') 
```

```python
# Obtain 3-dimensional spaces
dim.new_dim = dict()
dim.apply_dim(num_dim=[3], tflag=[3, 6, 7, 8])

# Plot spaces
dim.plot_artificial_3D(n_rows=2, n_cols=4, figsize=(15, 8), save_name='artificial_3d_24')    
```

# Multilabel Artificial Dataset

```python
# Create and split multilabel dataset
X, y = make_multilabel_classification(150, n_features=300, n_classes=2, random_state=34)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=33)

# Initialize dim. red. framework
dim = Dim()
dim.X_train = X_train
dim.y_train = y_train
dim.X_test = X_test
dim.y_test = y_test
```

```python
# Obtain 2-dimensional spaces
dim.new_dim = dict()
dim.apply_dim(num_dim=2, tflag=[6, 9], multilabel=True)

# Plot spaces in 2D
dim.plot_artificial_multilabel(n_rows=3, n_cols=3, figsize=(12, 12), save_name='2D-multilabel')
```
